```python
%%capture

import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
import math
import os

plt.tight_layout()
plt.style.use('ggplot')
plt.rcParams.update({'font.size': 18, 'font.family': 'sans-serif'})
```

```python
colNames = [
    '\$0 - $30,000 a year',
    '\$30,001 - $48,000 families',
    '\$48,001 - $75,000 families',
    '\$75,001 - $110,000 families',
    'Average net price for families earning \$110,000+ families',
]
data = [
    [4676.0, 2808.0, 3768.91, 4173.0],
    [5307.0, 4157.0, 4563.29, 4906.18],
    [9210.0, 8069.0, 6897.21, 6652.98],
    [17143.0, 16018.0, 14713.46, 15374.21],
    [38968.0, 37520.0, 40437.21, 35963.83],
]
years = [2013, 2014, 2015, 2016]
```

```python
plt.figure(figsize=(16,16))
for i in range(len(data)-1, -1, -1):
    plt.plot(years, data[i], label=colNames[i], linewidth=3)
plt.xlabel('Year (autumn)', labelpad=5)
plt.ylabel('Inflation-adjusted dollars')
plt.title('Average net prices for CC and SEAS only', y=1.01)
plt.xticks(np.arange(years[0], years[-1] + 1, 1))
plt.legend(prop={'size': 18})
plt.savefig('../img/anp-ccseas.png', bbox_inches='tight')
```
